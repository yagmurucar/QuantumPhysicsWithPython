```python
import numpy as np
import matplotlib.pyplot as plt

# rnek bir potansiyel enerji fonksiyonu . (rnein, harmonik osilatr: V(x) = (1/2) * k * x^2)
k = 1.0  # Yay sabiti
E = 5.0  # Toplam enerji 

# x araln belirle
x = np.linspace(-10, 10, 400)  # x deerleri, -10 ile 10 arasnda 400 nokta

# Potansiyel enerji U(x) = V(x)
U = 0.5 * k * x**2

# Kinetik enerji K(x) = E - U(x)
K = E - U

# Grafik izimi
plt.figure(figsize=(10, 6))

# Potansiyel enerji U(x)
plt.plot(x, U, label='U(x) = Potansiyel Enerji', color='green')

# Kinetik enerji K(x)
plt.plot(x, K, label='K(x) = E - U(x)', color='black', linestyle='--')

# Toplam enerji E (yatay izgi)
plt.axhline(y=E, color='yellow', linestyle='-', label='E = Toplam Enerji')

# Klasik dnm noktalar (K=0 olan yerler)
turning_points = np.where(np.isclose(K, 0, atol=1e-2))[0]
if len(turning_points) > 0:
    plt.scatter(x[turning_points], U[turning_points], color='black', label='Dnm Noktalar')

# Eksen etiketleri ve balk
plt.xlabel('x (Konum)')
plt.ylabel('Enerji')
plt.title('Zamandan Bamsz Schrdinger Denklemi: U(x) + K(x) = E')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametreler
k = 1.0          # Harmonik osilatr sabiti
E = 7.5          # stediin enerji seviyesi 

# x aral
x = np.linspace(-12, 12, 600)

# Potansiyel ve kinetik enerji
U = 0.5 * k * x**2
K = E - U

plt.figure(figsize=(11, 7))

plt.plot(x, U, color='#FF6B00', linewidth=4, label='U(x) = Potansiyel Enerji')

plt.plot(x, K, color='#FFD93D', linewidth=4, linestyle='--', 
         label='K(x) = E - U(x)  (kesikli)')

plt.axhline(y=E, color='black', linewidth=4, linestyle='-', 
            label=f'E = {E}  (sabit toplam enerji)')

turning_x = x[np.abs(K) < 0.03]
plt.scatter(turning_x, np.full_like(turning_x, E), 
            color='#FF4500', s=120, zorder=5, edgecolors='black', linewidth=1.5,
            label='Klasik Dnm Noktalar')

plt.xlabel('x (konum)', fontsize=14, fontweight='bold')
plt.ylabel('Enerji', fontsize=14, fontweight='bold')
plt.title('Zaman Bamsz Schrdinger Denklemi\nU(x) + K(x) = E', 
          fontsize=16, fontweight='bold', pad=20)

plt.legend(fontsize=12, loc='upper center', fancybox=True, shadow=True)
plt.grid(True, alpha=0.3)
plt.xlim(-12, 12)
plt.ylim(-1, 18)

plt.gca().set_facecolor('#FFFAF0')

plt.tight_layout()
plt.show()
```
